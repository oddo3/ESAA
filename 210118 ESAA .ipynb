```python
from pandas import DataFrame

#예제 데이터 만들기

ex = {'경도' : [126.94477659259302, 126.9435395978225, 126.94185781346485, 126.94488904200215,126.94493195734583],
     '위도' : [37.55866478781588, 37.559133965150224, 37.55738454631857, 37.55970734855624, 37.55826999866026],
     '상호명' : ['삭', '반서울', '방콕익스프레스', '리또리또', '다다']}

 
ex=DataFrame(ex)
ex
```

```python
import folium

#지도의 중심을 지정하기 위해 위도와 경도의 평균 구하기
lat = ex['위도'].mean()
long = ex['경도'].mean()

#지도 띄우기
m = folium.Map([lat,long],zoom_start=16.5)
m
```

```python
for i in ex.index:
    sub_lat =  ex.loc[i,'위도']
    sub_long = ex.loc[i,'경도']
    
    title = ex.loc[i,'상호명']
    
    #지도에 데이터 찍어서 보여주기
    folium.Marker([sub_lat,sub_long],tooltip = title).add_to(m)

#한글이 안나오는 오류로 html로 trouble shooting 
m.save('example.html')
m
```

```python
#지도 새로 띄우기
m = folium.Map([lat,long],zoom_start=16.5)

for i in ex.index:
    sub_lat =  ex.loc[i,'위도']
    sub_long = ex.loc[i,'경도']
    
    title = ex.loc[i,'상호명']
    
    color = 'green'
    if ex.loc[i,'상호명'] == '삭':
        color = "red"
        
    #지도에 동그라미로 데이터 찍기    
    folium.CircleMarker([sub_lat,sub_long],color=color,radius = 5, tooltip=title).add_to(m)

#한글이 안나오는 오류로 html로 trouble shooting
m.save('example.html')
m
```
